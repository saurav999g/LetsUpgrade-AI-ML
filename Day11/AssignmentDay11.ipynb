```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
dataset = pd.read_csv('general_data.csv')
```

```python
dataset
```

```python
dataset.duplicated()
```

```python
dataset.drop_duplicates()
```

```python
dataset['Attrition'] = dataset['Attrition'].map({'Yes' : 1, 'No' : 0})
```

```python
dataset
```

```python
stats , p = pearsonr( dataset['Attrition'], dataset['Age'])
print( stats, p)
```

Age is weakly negatively correlated to Attrition. 

```python
stats , p = pearsonr( dataset['Attrition'], dataset['DistanceFromHome'])
print( stats, p)
```

p value is greter than 0.05 hence H0 will be accepted

```python
plt.scatter(dataset['Attrition'], dataset['DistanceFromHome'])
```

```python
stats1 , p1 = pearsonr( dataset['Attrition'], dataset['Education'])
print( stats1, p1)
```

Education is very weakly negatively correlated to Attrition and p value is greter then 0.05 hence H0 accepted.
