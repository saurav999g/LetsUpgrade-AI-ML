Chi square test (for categorical variables only)

```python
from scipy.stats import chi2_contingency
```

```python
chitable = pd.crosstab(dataset1.Attrition, dataset1.BusinessTravel)
print(chitable)
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
print(stats, p)
```

As p value is less than 0.05 reject H0 (Null hypothesis)<br>
H0 - There is no dependency between Attrition and Business Travel.
Ha - There is dependency between Attrition and Business Travel

```python
chitable = pd.crosstab(dataset1.Attrition, dataset1.Department)
print(chitable)
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
print(stats, p)
```

Here also the p value is less than 0.05 hence null hypothesis will be rejected.<br>
Ha - There is dependency between Department and Attrition rate. 

```python
chitable = pd.crosstab(dataset1.Attrition, dataset1.EducationField)
print(chitable)
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
print(stats, p)
```

Mann whitney ( Two independent samples)

```python
from scipy.stats import mannwhitneyu
```

```python
chitable = pd.crosstab( dataset1.Age,dataset1.Attrition)
chitable
```

```python
stats, p = mannwhitneyu(chitable.No, chitable.Yes)
```

```python
print(stats,p)
```

H0 - There is no relation between age and attrition.<br>
Ha - There is relation between age and attrition.<br>
as p < 0.05 h0 rejected

```python
chitable = pd.crosstab(dataset1.PercentSalaryHike, dataset1.Attrition)
print(chitable)
```

```python
stats,p = mannwhitneyu(chitable.No,chitable.Yes)
print(stats,p)
```

H0 - There is no difference between Percentsalaryhike of person who didn't left the company and the one who left the company.<br>
Ha - There is significant  difference between Percentsalaryhike of person who didn't left the company and the one who left the company.<br>
Here in this case the p value is less than 0.05 hence H0 will be rejected.
